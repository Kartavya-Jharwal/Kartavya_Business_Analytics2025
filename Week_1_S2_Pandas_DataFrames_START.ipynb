```python
# @title
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Generate e-commerce dataset
np.random.seed(42)  # For reproducibility
n_customers = 1000

# Generate customer data
customer_data = {
    'customer_id': range(1, n_customers + 1),
    'age': np.random.normal(35, 12, n_customers).astype(int),
    'total_purchases': np.random.randint(1, 50, n_customers),
    'average_order_value': np.random.normal(75, 25, n_customers),
    'membership_months': np.random.randint(1, 60, n_customers),
    'product_categories': np.random.choice(['Electronics', 'Clothing', 'Books', 'Home', 'Sports'], n_customers),
    'is_premium_member': np.random.choice([True, False], n_customers, p=[0.3, 0.7])
}

df_ecommerce = pd.DataFrame(customer_data)
df_ecommerce['total_spent'] = df_ecommerce['total_purchases'] * df_ecommerce['average_order_value']
```

```python
# @title code for student dataset
# Generate student performance dataset
n_students = 500

# Generate grades with some realistic correlations
base_ability = np.random.normal(75, 10, n_students)
study_hours = np.random.normal(15, 5, n_students)

student_data = {
    'student_id': range(1, n_students + 1),
    'study_hours_per_week': np.maximum(0, study_hours),
    'math_score': np.clip(base_ability + np.random.normal(0, 5, n_students) + (study_hours - 15) * 0.5, 0, 100),
    'science_score': np.clip(base_ability + np.random.normal(0, 5, n_students) + (study_hours - 15) * 0.4, 0, 100),
    'english_score': np.clip(base_ability + np.random.normal(0, 7, n_students) + (study_hours - 15) * 0.3, 0, 100),
    'attendance_rate': np.clip(np.random.normal(0.9, 0.05, n_students), 0, 1),
    'extracurricular_activities': np.random.randint(0, 4, n_students)
}

df_students = pd.DataFrame(student_data)
```

### Exercise 2: Student Performance Analysis

1. Statistical Analysis:
   - Calculate mean, median, and standard deviation for each subject
   - Find the correlation between study_hours and scores
   - How does attendance_rate affect performance?

2. Performance Metrics:
   - Create an overall performance score
   - Identify top 10% performers
   - Analyze impact of extracurricular activities on scores




```python
df_students.head()
```

**Statistical Analysis:**



```python
# Basic statistics for each subject
print(df_students['math_score'].describe())
print(df_students['science_score'].describe())
print(df_students['english_score'].describe())
print(df_students['extracurricular_activities'].describe())
```

```python
# Correlation between study hours and scores
df_students.corr()
```

```python
# Create attendance 4 groups
df_students['attendance_group'] = pd.cut(df_students['attendance_rate'], bins=[0, 0.5, 0.75, 1], labels=['Low', 'Medium', 'High'])
print(df_students['attendance_group'].describe())
```

```python
df_students.head()
```

**Performance Metrics**

```python
# Overall performance score
df_students['overall_score'] = df_students[['math_score', 'science_score', 'english_score']].mean(axis=1)
print(df_students['overall_score'].describe())
```

```python
# Extracuricular impact
top_10_percent = df_students[df_students['overall_score'] >= df_students['overall_score'].quantile(0.9)]
print(top_10_percent['extracurricular_activities'].describe())
```
